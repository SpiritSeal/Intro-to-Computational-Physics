```python
import matplotlib.pyplot as plt
import numpy as np
```

This imports the matplotlib.pyplot and numpy libraries, which are necessary for creating plots and generating data in Python.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

We create some data `x` and `y` using numpy function `linspace()` and `sin()`. `linspace(0,10,100)` creates an array of 100 evenly spaced numbers over the range of 0 to 10, and `sin(x)` gives the sine of each element of the array.

```python
plt.plot(x, y)
```

This plots the data on the x-axis and y-axis respectively.

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('A sine wave')
```

This adds axis labels and a title to the plot.

```python
plt.show()
```

This last line is used to display the plot.

It is a basic example of how to create a line plot using matplotlib in python. There are many other types of plots that can be created using matplotlib such as scatter plot, bar plot, histogram, etc. And also many other customization options available to make the plot more informative and visually appealing.
