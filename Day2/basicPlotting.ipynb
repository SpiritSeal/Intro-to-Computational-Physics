```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

This imports the matplotlib.pyplot and numpy libraries, which are necessary for creating plots and generating data in Python.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

We create some data `x` and `y` using numpy function `linspace()` and `sin()`. `linspace(0,10,100)` creates an array of 100 evenly spaced numbers over the range of 0 to 10, and `sin(x)` gives the sine of each element of the array.

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('A sine wave')
```

This is a basic example of how to create a line plot using matplotlib in python. ## Selected Examples

### Colored Bar Chart

```python
fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
# 'tab' a more formal-looking color scheme. Compare tab:red and red.
bar_colors = ['tab:red', 'tab:blue', 'red', 'tab:orange']

# For optional parameters, we can use keyword arguments. (basically: "nameOfParameter=value"; see 'color' below)
ax.bar(fruits, counts, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
```

### Grouped Bar Chart with Labels

Think of labels as the names of the bars.

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()
```

### Finding a line of best fit

```python
#define data
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([2, 5, 6, 7, 9, 12, 16, 19])

#find line of best fit (more on this later)
a, b = np.polyfit(x, y, 1)

#add points to plot
plt.scatter(x, y)

#add line of best fit to plot
plt.plot(x, a*x+b)    
```
