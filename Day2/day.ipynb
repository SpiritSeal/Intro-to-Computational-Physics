# How to use Python to plot data

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Numpy

Numpy is a Python library that provides a lot of useful functions for working with data.

```python
# Creating a 11 evenly spaced numbers between 0 and 10 inclusive
np.linspace(0, 10, 11)
```

```python
# Plotting data using Python

def example_plot():
    # Creating a numpy array of 100 evenly spaced numbers between 0 and 10 inclusive
    # (To be used as the x-axis)
    x = np.linspace(0, 10, 100)

    # Plotting a 
    plt.plot(x, np.sin(x), '-')
    plt.plot(x, np.cos(x), '--')

    # Scaling the plot (Not including this line will result in a plot with an adaptive scale)
    plt.axis('equal')

    # Adding labels
    plt.xlabel("x")
    plt.ylabel("sin(x) and cos(x)")

    # Adding title
    plt.title("Plotting sin(x) and cos(x)")

    # Adding legend
    plt.legend(["sin(x)", "cos(x)"])

    # Displaying the plot
    plt.show()
```

```python
example_plot()
```

You can use the Interact Library to create interactive plots.

```python
# interactive plotting import
import ipywidgets as widgets
```

```python
# Plot x vs sin(x) and a*sin(b*x+c)+d
def interactive_plot(a, b, c, d):
    # Numpy array
    x = np.linspace(0, 10, 100)

    # Plotting the data
    plt.plot(x, np.sin(x), '-')
    plt.plot(x, a*np.sin(b*x+c)+d, '--')

    # Scaling the plot (Not including this line will result in a plot with an adaptive scale)
    plt.axis('equal')

    # Adding labels
    plt.xlabel("x")
    # plt.ylabel("sin(x) and a*sin(b*x+c)+d")
    plt.ylabel(f"{a}*sin({b}*x+{c})+{d}")

    # Adding title
    plt.title(f"Plotting sin(x) and {a}*sin({b}*x+{c})+{d}")

    # Adding legend

    plt.legend(["sin(x)", f"{a}*sin({b}*x+{c})+{d}"])

    # Displaying the plot
    plt.show()
```

```python
# Creating interactive plot slider
widgets.interact(interactive_plot, 
a=widgets.FloatSlider(min=0, max=2, step=0.1, value=1),
b=widgets.FloatSlider(min=0, max=2, step=0.1, value=1),
c=widgets.FloatSlider(min=-1, max=1, step=0.1, value=0),
d=widgets.FloatSlider(min=-1, max=1, step=0.1, value=0))
```
